```python
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco
        
    def etiqueta(self):
        return Panel (f'[white]{self.nome}\n {"-" * 30} \n ----------R${self.preco}----------[/]', title='Produto', style='white', width=38)
```

```python
p1 = Produto('iPhone 17 Pro Max', 25000.85)
p1.etiqueta()
```
