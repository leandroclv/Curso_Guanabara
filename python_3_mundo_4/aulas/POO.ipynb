###Criando classes e objetos na prÃ¡tica

```python
#DeclaraÃ§Ã£o de classe
class Gafanhoto:
    def __init__(self): #MÃ©todo construtor
        #Atributo de instÃ¢ncia
        self.nome = ''
        self.idade = 0
        
    #MÃ©todo de instÃ¢ncia
    def aniversario(self):
        self.idade += 1
        
    def mensagem(self):
        return f'{self.nome} Ã© Gafanhoto(a) e tem {self.idade} anos de idade'
```

```python
#DeclaraÃ§Ã£o de Objetos
g1 = Gafanhoto()
g1.nome = 'Maria'
g1.idade = 17
g1.mensagem()
```

```python
g1.aniversario()
print(g1.mensagem())
```

```python
g2 = Gafanhoto()
g2.nome = 'Mauro'
g2.idade = 53
g2.mensagem()
```

```python
g3 = Gafanhoto()
g3.mensagem()
```

```python
print(g1.mensagem())
print(g2.mensagem())
print(g3.mensagem())
```

###Melhorando classes e criando uma conta bancÃ¡ria na prÃ¡tica

```python
#DeclaraÃ§Ã£o de classe
class Gafanhoto:
    #Docstring -> documentaÃ§Ã£o 
    '''
    Essa classe cria um Gafanhoto, que Ã© uma pessoa que tem nome e idade.
    Para criar uma nova pessoa use 
    variÃ¡vel = Gafanhoto(nome, idade)
    '''
    def __init__(self, n = '', i = 0): #MÃ©todo construtor
        #Atributo de instÃ¢ncia
        self.nome = n
        self.idade = i
        
    #MÃ©todo de instÃ¢ncia
    def aniversario(self):
        self.idade += 1
        
    #def mensagem(self):
    #    return f'{self.nome} Ã© Gafanhoto(a) e tem {self.idade} anos de idade'
    
    def __str__(self): #Dunder Method
        return f'{self.nome} Ã© Gafanhoto(a) e tem {self.idade} anos de idade'
    
    def __getstate__(self):
        return f'Estado: nome = {self.nome} ; idade = {self.idade}'
```

```python
g1 = Gafanhoto('Maria', 17)
#g1.mensagem()
print(g1)
```

```python
g2 = Gafanhoto('Mauro', 53)
#g2.mensagem()
print(g2)
```

```python
g1.aniversario()
#g1.mensagem()
print(g1)
```

```python
print(g1.__doc__) #Dunder Attribute
```

```python
#print(g1) -> <__main__.Gafanhoto at 0x7ae2c428fa10> Essa Ã© a mensagem original do __str__ mostrando o local desta variÃ¡vel na memÃ³ria
print(g1) #-> Maria Ã© Gafanhoto(a) e tem 18 anos de idade Essa Ã© a mensagem do __str__ depois que foi trocado pelo mensagem no mÃ©todo de instÃ¢ncia
```

```python
print(g1.__dict__) #Attribute
```

```python
print(g1.__getstate__()) #Method
```

```python
print(g1.__class__)
```
