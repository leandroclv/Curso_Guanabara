###Criando classes e objetos na prática

```python
#Declaração de classe
class Gafanhoto:
    def __init__(self): #Método construtor
        #Atributo de instância
        self.nome = ''
        self.idade = 0
        
    #Método de instância
    def aniversario(self):
        self.idade += 1
        
    def mensagem(self):
        return f'{self.nome} é Gafanhoto(a) e tem {self.idade} anos de idade'
```

```python
#Declaração de Objetos
g1 = Gafanhoto()
g1.nome = 'Maria'
g1.idade = 17
g1.mensagem()
```

```python
g1.aniversario()
print(g1.mensagem())
```

```python
g2 = Gafanhoto()
g2.nome = 'Mauro'
g2.idade = 53
g2.mensagem()
```

```python
g3 = Gafanhoto()
g3.mensagem()
```

```python
print(g1.mensagem())
print(g2.mensagem())
print(g3.mensagem())
```

###Melhorando classes e criando uma conta bancária na prática

```python
#Declaração de classe
class Gafanhoto:
    #Docstring -> documentação 
    '''
    Essa classe cria um Gafanhoto, que é uma pessoa que tem nome e idade.
    Para criar uma nova pessoa use 
    variável = Gafanhoto(nome, idade)
    '''
    def __init__(self, n = '', i = 0): #Método construtor
        #Atributo de instância
        self.nome = n
        self.idade = i
        
    #Método de instância
    def aniversario(self):
        self.idade += 1
        
    #def mensagem(self):
    #    return f'{self.nome} é Gafanhoto(a) e tem {self.idade} anos de idade'
    
    def __str__(self): #Dunder Method
        return f'{self.nome} é Gafanhoto(a) e tem {self.idade} anos de idade'
    
    def __getstate__(self):
        return f'Estado: nome = {self.nome} ; idade = {self.idade}'
```

```python
g1 = Gafanhoto('Maria', 17)
#g1.mensagem()
print(g1)
```

```python
g2 = Gafanhoto('Mauro', 53)
#g2.mensagem()
print(g2)
```

```python
g1.aniversario()
#g1.mensagem()
print(g1)
```

```python
print(g1.__doc__) #Dunder Attribute
```

```python
#print(g1) -> <__main__.Gafanhoto at 0x7ae2c428fa10> Essa é a mensagem original do __str__ mostrando o local desta variável na memória
print(g1) #-> Maria é Gafanhoto(a) e tem 18 anos de idade Essa é a mensagem do __str__ depois que foi trocado pelo mensagem no método de instância
```

```python
print(g1.__dict__) #Attribute
```

```python
print(g1.__getstate__()) #Method
```
